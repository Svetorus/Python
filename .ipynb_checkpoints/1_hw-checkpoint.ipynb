### 4. 
Преобразовать слова «разработка», «администрирование», «protocol», «standard» из строкового представления в байтовое и выполнить обратное преобразование (используя методы encode и decode).

```python
a4 = ('разработка')
b4 = ('администрирование')
c4 = ('protocol')
d4 = ('standart')
```

```python
type(a4),type(b4),type(c4),type(d4)
```

```python
a4_bytes = a4.encode()
b4_bytes = b4.encode()
c4_bytes = c4.encode()
d4_bytes = d4.encode()
```

```python
a4_bytes
```

```python
type(a4_bytes),type(b4_bytes),type(c4_bytes),type(d4_bytes)
```

```python
a41 = a4_bytes.decode()
b41 = b4_bytes.decode()
c41 = c4_bytes.decode()
d41 = d4_bytes.decode()
```

```python
a41
```

```python
type(a41),type(b41),type(c41),type(d41)
```
